```python
import sklearn
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = datasets.load_iris()
```

```python
data.target_names
```

```python
df = pd.DataFrame(
    data=data.data,
    columns=["_".join(x.split(" ")[:2]) for x in data.feature_names],
)
df.head()
```

```python
df["target"] = data.target
df["target_name"] =[data.target_names[x] for x in df["target"]]
df
```

```python
df["target_name"].value_counts().plot(kind = "bar")
```

```python
df.plot(kind = "scatter", x = "petal_width", y = "sepal_width")
```

```python
df["petal_length"].plot(kind = "kde")
```

```python
sns.pairplot(df, hue="target_name",)
```
